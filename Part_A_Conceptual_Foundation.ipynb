### 1.1 What is Data Analysis?

Data Analysis is the process of collecting, cleaning, transforming, and interpreting data to discover meaningful insights, patterns, and trends. It helps organizations and individuals make informed decisions based on evidence rather than assumptions.

The main goal of data analysis is to convert raw data into useful information. In modern industries such as finance, healthcare, marketing, education, and technology, data analysis plays a critical role in improving efficiency and solving problems.

#### Steps in Data Analysis
1. Data Collection  
   ##### Gathering data from sources such as databases, APIs, surveys, sensors, websites, or CSV files.

2. Data Cleaning
   ##### Removing missing values, duplicates, and inconsistencies to improve data quality.

3. Data Transformation 
   ##### Converting data into a suitable format for analysis using normalization, encoding, scaling, etc.

4. Exploratory Data Analysis (EDA) 
   ##### Understanding patterns using statistics and visualizations.

5. Modeling and Interpretation  
   ##### Applying statistical or machine learning techniques to derive conclusions.

6. Communication of Results  
   ##### Presenting findings using charts, dashboards, or reports.

#### Types of Data Analysis
- Descriptive Analysis – Explains what happened.
- Diagnostic Analysis – Explains why it happened.
- Predictive Analysis – Predicts future outcomes.
- Prescriptive Analysis – Suggests actions to take.

#### Importance of Data Analysis
- Helps businesses make data-driven decisions.
- Identifies customer behavior and trends.
- Improves operational efficiency.
- Reduces risks and costs.
- Supports machine learning and AI systems.

### 1.2 How to Plan a Data Science Project

Planning is one of the most important stages of a data science project because it ensures that the project has a clear objective, proper workflow, and measurable outcomes.

A successful data science project usually follows a structured lifecycle.

#### Steps to Plan a Data Science Project

### 1. Define the Problem
The first step is understanding the business or research problem clearly.  
Example:
- Predict customer loan default.
- Detect fraud transactions.
- Forecast sales.

A good problem statement should be:
- Specific
- Measurable
- Realistic
- Goal-oriented

### 2. Understand Requirements
Identify:
- Required data sources
- Tools and technologies
- Expected outputs
- Project timeline
- Team responsibilities

### 3. Data Collection
Collect relevant data from:
- CSV files
- Databases
- APIs
- Web scraping
- Sensors or logs

The quality and quantity of data directly affect model performance.

### 4. Data Preparation
This stage includes:
- Handling missing values
- Removing duplicates
- Feature engineering
- Data encoding and scaling
- Splitting data into training and testing sets

Data preparation usually consumes the majority of project time.

### 5. Exploratory Data Analysis (EDA)
EDA helps understand:
- Data distributions
- Correlations
- Outliers
- Patterns and trends

Visualization libraries such as Matplotlib and Seaborn are commonly used.

### 6. Model Selection and Training
Choose suitable machine learning algorithms depending on the problem:
- Regression for continuous predictions
- Classification for categorical predictions
- Clustering for grouping data

Train the model using historical data.

### 7. Model Evaluation
Evaluate model performance using metrics such as:
- Accuracy
- Precision
- Recall
- F1-score
- RMSE
- MAE

The evaluation determines whether the model meets project goals.

### 8. Deployment
Deploy the model into a real-world environment using:
- Web applications
- APIs
- Cloud platforms

### 9. Monitoring and Maintenance
After deployment, models should be monitored regularly because:
- Data patterns may change over time.
- Model performance can decrease.
- Retraining may be required.

#### Importance of Project Planning
- Reduces project failure risk
- Improves workflow management
- Saves time and resources
- Ensures better communication among team members

### 1.3 How to Frame a Machine Learning Problem

Framing a machine learning problem means converting a real-world problem into a format that machine learning algorithms can solve effectively.

Proper problem framing is extremely important because the success of the model depends on how clearly the problem is defined.

#### Steps to Frame a Machine Learning Problem

The first step is identifying what the organization wants to achieve.

Examples:
- Reduce loan defaults
- Predict customer churn
- Recommend products
- Detect spam emails

The target variable is the output the model needs to predict.

Examples:
- Fraud = Yes/No
- House Price
- Customer Churn

### 3. Identify Input Features
Features are the variables used to make predictions.

Examples:
- Age
- Income
- Transaction Amount
- Credit Score

Feature quality strongly impacts model accuracy.

### 4. Determine the Type of Problem

#### Classification
Predicts categories or labels.
Examples:
- Spam detection
- Disease prediction

#### Regression
Predicts continuous numerical values.
Examples:
- Stock prices
- House prices

#### Clustering
Groups similar data points together.
Examples:
- Customer segmentation

### 5. Select Evaluation Metrics
Metrics depend on the problem type.

#### Classification Metrics

#### Regression Metrics


### 6. Check Data Availability
A machine learning system requires sufficient, relevant, and clean data.

Questions to consider:
- Is enough data available?

- Are labels accurate?

### 7. Identify Constraints
Constraints may include:
- Computational cost
- Time limitations
- Privacy regulations
- Real-time prediction requirements

Clearly define what success means.

Example:

- Reduce prediction error below a certain threshold.

# 2. Tensors with NumPy Examples

## What is a Tensor?

A Tensor is a mathematical structure used to represent numerical data in multiple dimensions. Tensors are the foundation of machine learning, deep learning, and scientific computing.

A tensor can be understood as a generalized form of arrays.

| Tensor Type | Dimensions | Example |

### | Scalar | 0D | 5 |
### | Vector | 1D | [1, 2, 3] |
### | Matrix | 2D | [[1,2],[3,4]] |
### | Higher-order Tensor | 3D+ | Multi-dimensional arrays |


NumPy arrays are commonly used to represent tensors in Python.

---

## Tensor Dimensions

### 0D Tensor (Scalar)

A scalar contains only one value.

```python
import numpy as np

scalar = np.array(10)
print(scalar)
print(scalar.ndim)
```

**Output**
```python
10
0
```

---

### 1D Tensor (Vector)

A vector is a one-dimensional array.

```python
vector = np.array([1, 2, 3, 4])
print(vector)
print(vector.ndim)
```

**Output**
```python
[1 2 3 4]
1
```

---

### 2D Tensor (Matrix)

A matrix contains rows and columns.

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(matrix)
print(matrix.ndim)
```

**Output**
```python
[[1 2 3]
 [4 5 6]]
2
```

---

### 3D Tensor

A 3D tensor contains multiple matrices.

```python
tensor_3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print(tensor_3d)
print(tensor_3d.ndim)
```

**Output**
```python
[[[1 2]
  [3 4]]

 [[5 6]
  [7 8]]]
3
```

---

## Important Tensor Properties

### Shape
The shape describes the dimensions of the tensor.

```python
print(matrix.shape)
```

**Output**
```python
(2, 3)
```

This means:
- 2 rows
- 3 columns

---

### Size
Returns the total number of elements.

```python
print(matrix.size)
```

**Output**
```python
6
```

---

### Data Type

```python
print(matrix.dtype)
```

**Output**
```python
int64
```

---

## Tensor Operations in NumPy

### Addition

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
```

**Output**
```python
[5 7 9]
```

---

### Multiplication

```python
print(a * b)
```

**Output**
```python
[ 4 10 18]
```

---

### Matrix Multiplication

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6], [7, 8]])

print(np.matmul(x, y))
```

**Output**
```python
[[19 22]
 [43 50]]
```

---

## Why Tensors are Important in Machine Learning

Tensors are heavily used in machine learning and deep learning because:

- They efficiently store numerical data.
- They support high-speed mathematical operations.
- Neural networks process inputs and outputs as tensors.
- Libraries such as TensorFlow and PyTorch are built around tensor operations.

Examples:
- Images are represented as 3D tensors.
- Videos are represented as 4D tensors.
- Text embeddings are stored as tensors.
